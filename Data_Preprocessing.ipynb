# The data is in jscon.gz,First defining a function getDF to get jscon file in dataframe format.


```python
import numpy as np
import pandas as pd
import gzip
import json
# Hide warnings
import warnings

warnings.simplefilter("ignore", UserWarning)
warnings.simplefilter("ignore", FutureWarning)
warnings.simplefilter("ignore", DeprecationWarning)
```

# Defining a function for exctracting information from json to dataframe.

```python
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index'
```

# Defining a function for checking null values.

```python
def cal_missing_value(df):
    total=df.isnull().sum().sort_values(ascending=False)
    percentage=(df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data=pd.concat([total,percentage], axis=1, keys=['Total','Percentage'])
    return missing_data
```

# Cleaning the review text.
Removing punctuation and special characters

```python
import re 
import string 
def reviewText_cleaning(text):
    
    
    text = str(text).lower()                                                      # Converting into smaller case.
    text = re.sub('https?://\S+|www\.\S+', '', text)                              # Removing urls links.
    text = re.sub('<.*?>+', '', text)                                             # Removing special characters.
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)               # Removing punctuation.
    text = re.sub('\n', '', text)                                                 # removing new lines.
    text = re.sub('\w*\d\w*', '', text)                                           # removing words containing numbers.
    return text
```

```python
Amazon_products1['reviewText']=Amazon_products1['reviewText'].apply(lambda x:reviewText_cleaning(x))
```

# About TextBlob
 TextBlob is a python library and offers a simple API to access its methods and perform basic NLP tasks. 

```python
from textblob import TextBlob
# create a function to get the subjectivity (opinionated level of text) and polarity -> 

def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
Amazon_products1['review_text']=Amazon_products1['review_text'].astype(str)
```

```python
#create a column for polarity:

Amazon_products1['polarity'] = Amazon_products1['review_text'].apply(getPolarity)
```

# To check if reviews are positive ,neutral and negative.Using the TextBlob for polarity on that basic we can check for what type of review it is.

```python

def GetAnalysis(score):
    if score <0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'

Amazon_products1['sentiment_Analysis'] = Amazon_products1['polarity'].apply(GetAnalysis)
```

```python
Amazon_products1['sentiment_Analysis'].value_counts()
```
